```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib as plt
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/data_hcm_final.csv')
df
```

```python
df2 = df[['ReviewUrl', 'LocationScore',	'PriceScore',	'QualityScore',	'ServingScore',	'SpaceScore']].dropna().drop_duplicates().reset_index(drop=True)
df2
```

# Outlier là khi 1 quán có các giá trị cách nhau 1 khoảng lớn hơn 5 điểm


```python
index_outliers = []

for i in range(len(df2)):
  if (abs(df2['PriceScore'][i] - df2['LocationScore'][i]) >= 5) or (abs(df2['PriceScore'][i] - df2['QualityScore'][i]) >= 5) or  (abs(df2['PriceScore'][i] - df2['ServingScore'][i]) >= 5) or  (abs(df2['PriceScore'][i] - df2['SpaceScore'][i]) >= 5):
    index_outliers.append(i)
for i in range(len(df2)):
  if i in index_outliers:
    df2 = df2.drop(i)

df2 = df2.reset_index(drop=True)
df2 = df2.rename(columns={"LocationScore": "Vị trí", "PriceScore": "Giá cả", "QualityScore": "Chất lượng", "ServingScore": "Phục vụ", "SpaceScore": "Không gian"})

df2
```

```python
import matplotlib.pyplot as plt

price = df2['Giá cả']
location = df2['Vị trí']
quality = df2['Chất lượng']
serving = df2['Phục vụ']
space = df2['Không gian']

fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(2, 2, figsize=(15, 15))
fig.suptitle("Mối quan hệ giữa yếu tố 'Giá cả' với 4 yếu tố còn lại", fontweight="bold", fontsize=20)
ax1.scatter(price, location)
ax1.set_title('Giá cả và Vị trí', fontsize=15)
ax1.set_xlabel('Giá cả', labelpad = 5)
ax1.set_ylabel('Vị trí', labelpad = 5)

ax2.scatter(price, quality)
ax2.set_title('Giá cả và Chất lượng', fontsize=15)
ax2.set_xlabel('Giá cả', labelpad = 5)
ax2.set_ylabel('Chất lượng', labelpad = 5)

ax3.scatter(price, serving)
ax3.set_title('Giá cả và Phục vụ', fontsize=15)
ax3.set_xlabel('Giá cả', labelpad = 5)
ax3.set_ylabel('Phục vụ', labelpad = 5)

ax4.scatter(price, space)
ax4.set_title('Giá cả và Không gian', fontsize=15)
ax4.set_xlabel('Giá cả', labelpad = 5)
ax4.set_ylabel('Không gian', labelpad = 5)
```

```python
import pandas as pd
from sklearn import linear_model
import statsmodels.api as sm

x = df2[['Vị trí', 'Chất lượng', 'Phục vụ',	'Không gian']]
y = df2['Giá cả']
 
# with sklearn
regr = linear_model.LinearRegression()
regr.fit(x, y)

print('Intercept: \n', regr.intercept_)
print('Coefficients: \n', regr.coef_)

# with statsmodels
x = sm.add_constant(x) # adding a constant
 
model = sm.OLS(y, x).fit()
predictions = model.predict(x) 
 
print_model = model.summary()
print(print_model)
```
